**Applying an aggregate value to each value you have grouped by**

```python
#Create a new column - 'max_order'

#Define what goes in it.
#first group by users
#then each user gets assisnged a value which is their maximum number of orders

df_subset['max_order'] = df_subset.groupby(['user_id'])['order_number'].transform(np.max)
```

What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

**Group by and check stats for those grouped values**

```python
#So we want to group by 'loyatly_flag' and then look at the stats of the 'prices'

df_subset.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

**Create a new column and define what goes in it**

```python
#Create a new column - 'average_order'

#Define what goes in it.
#first group by users
#then each user gets assigned a value which is their average order price

df_subset['average_order'] = df_subset.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Use loc() to set up an if-else condition and to name different tupes of spenders based on the average price of products they buy

df_subset.loc[df_subset['average_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_subset.loc[df_subset['average_order'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#checking how many of each type of customer this produces

df_subset['spending_flag'].value_counts(dropna = False)
```

```python
#checking the sum of the spending flags adds up to 1 million

df_subset['spending_flag'].value_counts(dropna = False).sum()
```

```python
#looking at head of only certain rows to check is doing  what should

df_subset[['user_id', 'spending_flag', 'average_order']].head(60)
```

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers.

Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


**2nd version - Create a new column and define what goes in it**

```python
#Create a new column - 'order_frequency'

#Define what goes in it.
#first group by users
#then each user gets assigned a value which is their median frequency of order

df_subset['order_frequency'] = df_subset.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Use loc() to set up an if-else condition and to name different tupes of users based on how frequently they buy

df_subset.loc[df_subset['order_frequency'] >20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_subset.loc[(df_subset['order_frequency'] <= 20) & (df_subset['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
df_subset.loc[df_subset['order_frequency'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
#checking how many of each type of customer this produces

df_subset['frequency_flag'].value_counts(dropna = False)
```

```python
#checking the sum of the frequency flags adds up to 1 million

df_subset['frequency_flag'].value_counts(dropna = False).sum()
```

```python
#looking at head of only certain rows to check is doing  what should

df_subset[['user_id', 'frequency_flag', 'order_frequency']].head(60)
```
