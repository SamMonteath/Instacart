Creating a new data frame

```python
# Create a data frame

df_test = pd.DataFrame()
```

```python
#Create a mixed type column

df_test['mix'] = ['a', 'b', 1, True]
```

```python
df_test.head()
```

Looking for mixed data types

```python
#Checking for mixed data types

for col in df_test.columns.tolist():
  weird = (df_test[[col]].applymap(type) != df_test[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_test[weird]) > 0:
    print (col)
```

```python
# Given the warning, I asked Co-Pilot to update the code

for col in df_test.columns:
    col_series = df_test[col]
    col_type = type(col_series.iloc[0])
    weird = col_series.map(type) != col_type
    if weird.any():
        print(col)
```

Changing data type

```python
#Setting the mix column as string type

df_test['mix'] = df_test['mix'].astype('str')
```

```python
##Checking what type it is

df_test['mix'].dtype
```

Looking for missing observations

```python
#Checking how many missing observations (cells/datapoints) there are in df_prods

df_prods.isnull().sum()
```

```python
##Note: much less useful if just do this
df_prods.isnull()
```

Create a dataframe of the null values in one column

```python
#So need to look at product_name only

#Create a new dataframe just of the null values in product_name, i.e. those where isnull() = True

df_nan = df_prods[df_prods['product_name'].isnull() == True]
```

```python
#Can look at this whole new data frame

df_nan
```

```python
#Finding the rows and columns

df_prods.shape
```

Create a data frame without the null values

```python
#Creates a new dataframe with all of the null values removed - i.e. removing the whole row

df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
#Finding the rows and columns

df_prods_clean.shape
```

```python
#Can see we've removed 16 rows - the result wanted
```

Finding duplicate rows

```python
# Finding dupplicates

## looking for rows that have identical data

df_dups = df_prods_clean[df_prods_clean.duplicated()]
```

```python
#seeing what we have

df_dups
```

```python
#Check shape first

df_prods_clean.shape
```

New dataframe without duplicates

```python
# then drop the duplicates, in a new dateframe

df_prods_clean_no_dups = df_prods_clean.drop_duplicates()
```

```python
#Check shape of new df

df_prods_clean_no_dups.shape
```

```python
#Export our clean data

df_prods_clean_no_dups.to_csv(os.path.join(r'/content/', 'products_checked.csv'))
```
