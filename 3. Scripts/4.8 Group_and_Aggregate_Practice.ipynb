**import pandas and a dataset**

```python
import pandas as pd
```

```python
data = {"Team": ["Red Sox", "Red Sox", "Red Sox", "Red Sox", "Red Sox", "Red Sox", "Yankees", "Yankees", "Yankees", "Yankees", "Yankees", "Yankees"],
		"Pos": ["Pitcher", "Pitcher", "Pitcher", "Not Pitcher", "Not Pitcher", "Not Pitcher", "Pitcher", "Pitcher", "Pitcher", "Not Pitcher", "Not Pitcher", "Not Pitcher"],
		"Age": [24, 28, 40, 22, 29, 33, 31, 26, 21, 36, 25, 31]}
df = pd.DataFrame(data)
print(df)
```

**Group by and show stats**

```python
df.groupby('Team').agg({'Age': ['mean', 'min', 'max']})
```

```python
# group by Team, get mean, min, and max value of Age for each value of Team.
grouped_single = df.groupby('Team').agg({'Age': ['mean', 'min', 'max']})

print(grouped_single)
```

**Renaming columns when showing the stats**

```python
# rename columns
grouped_single.columns = ['age_mean', 'age_min', 'age_max']

# reset index to get grouped columns back
grouped_single = grouped_single.reset_index()

print(grouped_single)
```

**Groupby more than one column**

```python
#Grouping by multiple columns

grouped_multiple = df.groupby(['Team', 'Pos']).agg({'Age': ['mean', 'min', 'max']})
grouped_multiple.columns = ['age_mean', 'age_min', 'age_max']
grouped_multiple = grouped_multiple.reset_index()
print(grouped_multiple)
```

**Import new data - setting data types**

```python
#Set data types

dtypes = {
    "first_name": "category",
    "gender": "category",
    "type": "category",
    "state": "category",
    "party": "category",
}
```

```python
#Import legislators .csv as df_leg

##This is saved in my top level of Section 4

#Only importing the columns we've specified above, and 'birthday' and 'last_name'

df_leg = pd.read_csv("legislators-historical.csv",
    dtype=dtypes,
    usecols=list(dtypes) + ["birthday", "last_name"],
    parse_dates=["birthday"]
)



```

**Groupby 2 columns ' 'sortby' count of another column**

```python
n_by_state_party = df_leg.groupby(["state", "party"])["last_name"].count()
```

**Create new column - use loc() and if-else to populate - with a default value**

```python
#Using loc() to set up an if-else condition and to categorise inot 3 different types of party

df_leg['3_parties'] = 'Other'  # default to Other


```

```python
df_leg.loc[df_leg['party'] == 'Republican', '3_parties'] = 'Republican'
df_leg.loc[df_leg['party'] == 'Democrat', '3_parties'] = 'Democrat'
```

```python
# Use boolean indexing to select rows where 'party' is not 'Democrat' or 'Republican'
## using this method - don't need this df_leg.loc[~df_leg['party'].isin(['Democrat', 'Republican']), '3_parties'] = 'Other'
```

```python
#Checking what the new column looks like

df_leg['3_parties'].value_counts()
```

```python
df_leg['3_parties'].value_counts().sum()
```

**Use new column for a new groupby**

```python
n_by_state_3party = df_leg.groupby(["state", "3_parties"])["last_name"].count()
```
