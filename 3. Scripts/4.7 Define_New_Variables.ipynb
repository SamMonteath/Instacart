**Creating a new function - a price flag - using if/elif/else**

```python
#Defining our function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

**Applying function - to a new'price_range' column**

```python
# Applying our function

df['price_range'] = df.apply(price_label, axis=1)
```

**Using if/elif/else to create a busy day flag - a NEW VARIABLE**

```python
#Defining our new variable - result - which is about busiest days

result = []

for value in df_ord_prod_id_comb['order_dow']:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#quick look

result
```

**Populating new column with created variable**

```python
#Populating a new column with our new varibale

df_ord_prod_id_comb['busiest_day'] = result
```

```python
#Checking frequency

df_ord_prod_id_comb['busiest_day'].value_counts(dropna=False)
```

```python
df_ord_prod_id_comb['order_dow'].value_counts(dropna=False)
```

**Amending our newly created 'busiest_day' column**

```python
#Changing the result to look at 2 busiest and 2 least busy days

result = []

for value in df_ord_prod_id_comb['order_dow']:
  if value in [0, 1]:
    result.append("Busiest days")
  elif value in [3, 4]:
    result.append("Least busy days")
  else:
    result.append("Regularly busy days")
```

**Changing column name**

```python
#Rename 'busiest day' to 'busiest days'

df_ord_prod_id_comb = df_ord_prod_id_comb.rename(columns={'busiest_day': 'busiest_days'})

```

**Putting new result into column**

```python
# Populating the 'busiest days' column with our result

df_ord_prod_id_comb['busiest_days'] = result
```

```python
#seeing how that has changed the column

df_ord_prod_id_comb['busiest_days'].value_counts(dropna=False)
```

```python
#Checking still get the same overall number of rows

df_ord_prod_id_comb['busiest_days'].value_counts(dropna=False).sum()
```

Same number again, so all good

**Creating a busiest times column**

**Find busiest times with .value_counts on a column**

```python
#What are the busiest times?

df_ord_prod_id_comb['order_hour_of_day'].value_counts(dropna=False)
```

**Create new variable 'result_hours' - using if/elif/else based on busiest times**

```python
#Defining new variable 'result_hours'

result_hours = []

for value in df_ord_prod_id_comb['order_hour_of_day']:
  if value in[10,11,14]:
    result_hours.append("Busiest time")
  elif value in [3,4,2]:
    result_hours.append("Least busy time")
  else:
    result_hours.append("Regularly busy time")
```

```python
# quick check

result_hours
```

**Create new column and attach variable**

```python
#creating a new column and attaching our new variable

df_ord_prod_id_comb['busiest_time'] = result_hours
```

```python
#Looking at the column

df_ord_prod_id_comb['busiest_time'].value_counts(dropna=False)
```
